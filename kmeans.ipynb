```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv(r"C:\Users\rshic\Downloads\Mall_Customers.csv")

# Extract the relevant features for clustering
features = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Choose the number of clusters (k)
k = 3  # You can adjust this based on your requirements

# Fit the KMeans model to the data
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(features)

# Visualize the clusters
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='rainbow')
plt.title('K-Means Clustering of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv(r"C:\Users\rshic\Downloads\Mall_Customers.csv")

# Exclude non-numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Display the first few rows of the dataset
print(numeric_df.head())

# Summary statistics
print(numeric_df.describe())

# Data types and missing values
print(numeric_df.info())

# Pair plot for numerical features
sns.pairplot(df, hue='Gender')
plt.title('Pair Plot of Numerical Features')
plt.show()

# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()

# Box plot for Annual Income and Spending Score by Gender
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Gender', y='Annual Income (k$)', data=df)
plt.title('Annual Income by Gender')

plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='Spending Score (1-100)', data=df)
plt.title('Spending Score by Gender')

plt.tight_layout()
plt.show()

# Correlation matrix
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
